```python
def actualizar_estado_editor(operaciones_usuario):
    texto_escrito = list()
    texto_actual = ""
    rehacer = list()

    for operacion in operaciones_usuario:
        if operacion == "DESHACER":
            rehacer.append(texto_actual)
            texto_actual = texto_escrito.pop()
        elif operacion == "REHACER":
            texto_escrito.append(texto_actual)
            texto_actual = rehacer.pop()
        else:
            texto_escrito.append(texto_actual)
            texto_actual = operacion
            rehacer = list()
    
    return "".join(texto_escrito) + texto_actual
```
